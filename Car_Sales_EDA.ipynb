## Exploratory Data Analysis

### Exploratory Data Analysis(EDA) is the method used to analyze and summarize data sets.It is the way of exploring data through visual summaries and graphics and identifies general patterns in the data. These patterns include outliers and features of the data that might be unexpected. EDA is an important first step in any data analysis.

### Outlier Detection from Numerical columns And Remove it using IQR Method

```python
sns.boxplot(x=data['Price'],palette='hls')
plt.show()
```

```python
sns.boxplot(x=data['Mileage'],palette='hls')
plt.show()
```

```python
sns.boxplot(x=data['EngV'],palette='hls')
plt.show()
```

```python
sns.boxplot(x=data['Year'],palette='hls')
plt.show()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data= data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data.shape
```
